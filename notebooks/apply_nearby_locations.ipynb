```python
import pandas as pd
import numpy as np
```

```python
df_nearby = pd.read_json("assets/location-cep-nearby.json", convert_axes=False, orient="index")
df_properties = pd.read_json("assets/olx_location_cep_price.json",orient="records", convert_dates=False)

df_nearby.head()
```

```python
import re

def find_related_nearby_location(cep):
    formatted_cep = re.sub('[^0-9]', '', cep)
    nearby_location = df_nearby[formatted_cep]["nearby_results"]
    return nearby_location

df_properties["caracteristicas"].head().apply(pd.Series)["cep"].apply(find_related_nearby_location)
```

```python
df_properties["nearby_locations"] = df_properties["caracteristicas"].apply(pd.Series)["cep"].apply(find_related_nearby_location)

df_properties["nearby_locations"]
```
